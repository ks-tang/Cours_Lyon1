```python
"""
MIF29 TP1

Voir exercice 79 du livret :
<http://mif29-crypto-sec.pages.univ-lyon1.fr/files/MIF29_CoursCrypto_2023.pdf>

A faire :

- Compléter le code
- Debugger
- Tester les propriétés d'homomorphie de Paillier

"""

from sympy import randprime, mod_inverse
from random import randint


def getprime():
    p = randprime(2 ** (1024 - 1), 2**1024)
    return p


def genkeys():
    p = getprime()
    q = getprime()
    while p == q:
        q = getprime()

    N = int(p * q)
    Phi = int(N - p - q + 1)
    return [N, mod_inverse(N, Phi)]


def encrypt(m, pk):
    # TODO compléter la définition de r
    r = randint(1,pk)
    N2 = pk * pk
    c = ((1 + m * pk) * pow(r, pk, N2)) % N2
    return int(c)


def decrypt(c, pk, sk):
    N2 = pk * pk
    # TODO compléter la définition de r
    r = pow(c,sk,pk)
    s = mod_inverse(r, pk)
    m = ((c * pow(s, pk, N2)) % N2 - 1) // pk
    return int(m)


################ Tests fonctions de chiffrement et déchiffrement ######################

pk, sk = genkeys()
print("pk = ", pk)
print("sk = ", sk)
print()


x = 2

X = encrypt(x, pk)

dx = decrypt(X, pk, sk)

print("Decrypt ( Encrypt (", x, ")) = ", dx)


################ Tests des propriétés d'homomorphie ####################################


def oplus(X,Y,pk):
  N2 = pk * pk
  Z = X*Y % N2
  return Z

def produitParConstante(X,y,pk):
   N2 = pk * pk
   Z = pow(X,y,N2)
   return Z


def oppose(X,pk):
  N2 = pk * pk
  Z = pow(X, -1, N2) 
  return Z

y = 1000
Y = encrypt(y, pk)

Z1=oplus(X,Y,pk)
Z2=produitParConstante(X,y,pk)
Z3=oppose(Y,pk)

#print(decrypt(Z1,pk,sk),decrypt(Z2,pk,sk),decrypt(Z3,pk,sk)-pk)
print(decrypt(Z1,pk,sk))
print(decrypt(Z2,pk,sk))
print(decrypt(Z3,pk,sk)-pk)
```
