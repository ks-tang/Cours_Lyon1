```python
day_count = (end_date - start_date).days + 1
for single_date in [d for d in (start_date + timedelta(n) for n in range(day_count)) if d <= end_date]:
    print(strftime("%Y-%m-%d", single_date.timetuple()))
```

```python
from datetime import date, timedelta

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = date(2013, 1, 1)
end_date = date(2015, 6, 2)
for single_date in daterange(start_date, end_date):
    print(single_date.strftime("%Y-%m-%d"))
```

## Bibliothèque Python de gestion de mots de passe

```python
from crypt import crypt, mksalt, METHOD_SHA512
from getpass import getpass
from hmac import compare_digest


def checker(plaintext, reference) -> bool:
    return compare_digest(crypt(plaintext, reference), reference)


if __name__ == "__main__":
    with open("sample-1.passwd", "rt", encoding="utf-8") as file:
        ref = file.readline().strip()
        pwd = getpass()
        print(checker(pwd, ref))
```

**Question 4.1 Avec une version récente de Python, un avertissement sur la dépréciation de crypt est affiché. Pourquoi ce module est-il déprécié ?**

Le module crypt est déprécié dans les versions récentes de Python car il utilise une ancienne méthode de hachage de mots de passe qui n'est pas considérée comme sécurisée. Le module crypt utilise la fonction de hachage cryptographique basée sur le chiffrement DES (Data Encryption Standard), qui est maintenant considérée comme obsolète en raison de ses limitations en termes de sécurité.

La méthode de hachage DES utilise une clé de chiffrement de 56 bits, ce qui la rend vulnérable aux attaques par force brute modernes. De plus, elle ne prend pas en charge des algorithmes de hachage plus sécurisés tels que SHA-256 ou SHA-512.

**Question 4.2 Pourquoi utilise-t’on deux fois reference dans la fonction checker() ?**

Dans la fonction checker(), le paramètre reference est utilisé deux fois pour effectuer la comparaison de manière sécurisée.

La fonction crypt(plaintext, reference) est utilisée pour hacher le mot de passe en clair plaintext en utilisant la même méthode de hachage que celle utilisée pour générer la référence du mot de passe stockée. Elle prend en compte le paramètre reference, qui représente la référence du mot de passe stockée.

Ensuite, la fonction compare_digest() est utilisée pour comparer le résultat du hachage du mot de passe en clair avec la référence du mot de passe stockée. Elle prend également en compte le paramètre reference. Cette fonction effectue une comparaison de manière sécurisée, ce qui signifie qu'elle évite les attaques par synchronisation et les fuites d'informations par le biais du temps d'exécution.

En utilisant deux fois le paramètre reference, on s'assure que le hachage du mot de passe en clair est effectué avec la même méthode et les mêmes paramètres que ceux utilisés pour générer la référence du mot de passe stockée. Cela garantit une comparaison précise et sécurisée entre les deux valeurs.

En résumé, le paramètre reference est utilisé deux fois dans la fonction checker() pour le hachage du mot de passe en clair et pour la comparaison sécurisée des résultats de hachage.

**Question 4.3 En quoi getpass() diffère-t’elle de input() ?**

La fonction getpass() est conçue spécifiquement pour l'obtention sécurisée de mots de passe ou d'autres informations sensibles. Lorsque vous utilisez getpass(), l'entrée de l'utilisateur est masquée à l'écran, ce qui signifie que les caractères saisis ne sont pas affichés. Cela garantit une meilleure sécurité et évite l'affichage du mot de passe ou d'autres informations sensibles à l'écran.

```python
from passlib.hash import pbkdf2_sha256
from getpass import getpass

def checker(plaintext, reference) -> bool:
    return pbkdf2_sha256.verify(plaintext, reference)

if __name__ == "__main__":
    with open("sample-1.passwd", "rt", encoding="utf-8") as file:
        ref = file.readline().strip()
        pwd = getpass()
        print(checker(pwd, ref))
```
