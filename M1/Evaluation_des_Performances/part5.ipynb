```python
file = open("out.nam", "r")

count_cbr = 0
count_tcp = 0
count_received_cbr = 0
count_received_tcp = 0

for line in file:
  line = line.split()

  if line[0] == '-' or line[0] == 'd' or line[0] == 'e':
    if line[18] == "{0.0":
      count_cbr += 1
    elif line[18] == "{1.0":
      count_tcp += 1

  if line[0] == 'r' or line[0] == '+':
    if line[18] == "{0.0":
      count_received_cbr += 1
    elif line[18] == "{1.0":
      count_received_tcp += 1

print("Paquets perdus CBR :", count_cbr)
print("Paquets perdus TCP :", count_tcp)

print("Paquets reçus CBR :", count_received_cbr)
print("Paquets reçus TCP :", count_received_tcp)

file.close()
```

```python
file = open("out.nam", "r")
packet_loss_count = {}

for line in file:
    if "d" in line.split()[0]:
        # Récupérer l'émetteur et le destinataire du paquet perdu
        sender = line.split()[18]
        receiver = line.split()[19]

        # Mettre à jour le nombre de paquets perdus pour l'émetteur
        if sender in packet_loss_count:
            packet_loss_count[sender] += 1
        else:
            packet_loss_count[sender] = 1

        # Mettre à jour le nombre de paquets perdus pour le destinataire
        if receiver in packet_loss_count:
            packet_loss_count[receiver] += 1
        else:
            packet_loss_count[receiver] = 1

file.close()

max_packet_loss = max(packet_loss_count.values())
for key, value in packet_loss_count.items():
    if value == max_packet_loss:
        print("Le flux avec le plus de paquets perdus est :", key)
        print(packet_loss_count)
        break


```
