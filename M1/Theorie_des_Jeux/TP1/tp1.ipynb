```python
import random
```

```python
def gain(i,j):
  if(i==j):
    return 5
  else:
    return -abs(i-j)
```

```python
def strategieJoueur1(historique, k):
    num = [2,3]
    tirage = random.choice(num)

    return tirage

def strategieJoueur2(historique, k):
    num = [1,4]
    tirage = random.choice(num)

    return tirage
```

```python
import numpy as np

def Test(n):
    gainJoueur1 = 0
    gainJoueur2 = 0
    historique = np.zeros((2,n-1))
    for k in range(n-1):
        i = strategieJoueur1(historique,k)
        j = strategieJoueur2(historique,k)
        historique[0,k] = i
        historique[1,k] = j
        gainJoueur1 = gainJoueur1 + gain(i,j)/n
        gainJoueur2 = gainJoueur2 - gain(i,j)/n

    print(historique)
    return gainJoueur1, gainJoueur2
```
