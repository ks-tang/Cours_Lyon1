```python
longitude = nx.get_node_attributes(g,"lon").values()
```

```python
latitude = nx.get_node_attributes(g,"lat").values()
```

```python
geo_dict = {}
for cle,lon,lat in zip(g.nodes, longitude, latitude):
  geo_dict[cle] = (lon,lat)
```

```python
import matplotlib.pyplot as plt
```

```python
degree_dict = dict(g.degree())
```

```python
cmap = plt.get_cmap("viridis")
```

```python
min_degree = min(degree_dict.values())
max_degree = max(degree_dict.values())
normalized_degrees = [(degree_dict[node] - min_degree) / (max_degree - min_degree) for node in g.nodes()]
```

```python
node_colors = [cmap(degree) for degree in normalized_degrees]
```

```python
nx.draw_networkx(g, pos=geo_dict, with_labels=False, node_size=1, width=0.05, node_color=normalized_degrees, cmap=cmap)
```

```python
pagerank_score = nx.pagerank(g)
```

```python
nx.set_node_attributes(g, pagerank_score, name='pagerank')
```

```python
import cdlib
```

```python
from cdlib import algorithms
```

```python
coms = algorithms.louvain(g)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
x = list(nx.get_node_attributes(g,'lon').values())
y = list(nx.get_node_attributes(g,'lat').values())
coms_dict = coms.to_node_community_map()
hues = list(coms_dict[n][0] for n in g.nodes())
plt.figure(figsize=(12,8))
sns.scatterplot(x=x, y=y, hue=hues, palette=sns.color_palette('tab20', len(coms.communities)), s=5)
```

```python
countries = nx.get_node_attributes(g, "country")
```

```python
partition = {}  # Create an empty dictionary to store the partition
for node, country in countries.items():
    if country not in partition:
        partition[country] = []  # Create a list for each unique country
    partition[country].append(node)
```

```python
partition_list = list(partition.values())
```

```python
nc = cdlib.NodeClustering(partition_list, g, "GroundTruth")
```

```python
from cdlib import evaluation
```

```python
ami_score = evaluation.adjusted_mutual_information(nc, coms)
print(ami_score)
```

```python
nmi_score = evaluation.normalized_mutual_information(nc, coms)
print(nmi_score)
```
